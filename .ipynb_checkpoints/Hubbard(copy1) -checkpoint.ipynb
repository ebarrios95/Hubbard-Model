```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Hamiltonian matrix
def makeH(N,u):
    t = 1.0
    H = np.zeros((N,N))
    for i in range(N-1):
        H[i,i+1] = -t 
        H[i+1,i] = -t
    E,A = np.linalg.eigh(H)
    E += (N-1)*u #Adding in the chemical potential term
    return E
```

```python
def partition(E,N):
    Z = np.zeros(100)
    for i in range(N):
        Z += np.exp(-E[i]/T)
    return Z

def ETOT(E,N):
    global T
    T = np.logspace(-1,1,100)
    e = np.zeros(100)
    for i in range(N):
        e += E[i]*(np.exp(-E[i]/T))
    Etot = (1/partition(E,N))*e
    return(Etot)
```

```python
#Chemical Potential Function and plots
N=np.arange(1,8)
u = .5
for i in range(len(N)): 
    E = makeH(N[i],u)
    ET = ETOT(E,N[i])
    plt.plot(T,ET/N[i])
plt.xlabel("Temperature")
plt.ylabel("$ET/N$")
plt.title("Non-interacting Hubbard Model")
```

```python
N = np.array([2,3,4,5,6,7,8,9,10,19,20])
ET = np.zeros((len(N),100))
for i in range(len(N)):
    E = makeH(N[i])
    #ET[i] = ETOT(E,N[i])
    ET = ETOT(E,N[i])
    plt.plot(T,ET/N[i])
    plt.xscale('log')
```

```python
E1_0 = ET[1] - ET[0]
E1_1 = ET[4] - ET[3]
E1_2 = ET[8] - ET[7]
E1_3 = ET[10] - ET[9]
plt.plot(T,E1_0,color ='YELLOW')
plt.plot(T,E1_1,color =  'GREEN')
plt.plot(T,E1_2)
plt.plot(T,E1_3,color = 'b')
plt.xscale('log')
```
