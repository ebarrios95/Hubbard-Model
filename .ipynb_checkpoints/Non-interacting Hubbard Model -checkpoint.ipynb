```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Hamiltonian matrix
def makeH(N,u):
    t = 1.0
    H = np.zeros((N,N))
    for i in range(N-1):
        H[i,i+1] = -t 
        H[i+1,i] = -t
    E,A = np.linalg.eigh(H)
    E += (N-1)*u #Adding in the chemical potential term
    return E
```

```python
def partition(E,N):
    Z = np.zeros(100)
    for i in range(N):
        Z += np.exp(-E[i]/T)
    return Z

def ETOT(E,N):
    global T
    T = np.logspace(-1,1,100)
    e = np.zeros(100)
    for i in range(N):
        e += E[i]*(np.exp(-E[i]/T))
    Etot = (1/partition(E,N))*e
    return(Etot)
```

```python
#Calling all functions to obtain plots
N=np.arange(1,8)
u = .1
ET = np.zeros((len(N),100))
plt.figure(figsize = (12,10))
for i in range(len(N)): 
    E = makeH(N[i],u)
    ET[i] = ETOT(E,N[i])
    plt.plot(T,ET[i]/N[i],label ="N = %d" %(N[i]))
    plt.xlabel("Temperature",size = 15)
    plt.ylabel("Energy per N sites",size = 15)
    plt.legend(loc = 'lower right', prop ={'size':15})
#plt.savefig('u = .1.png')
```

```python
plt.figure(figsize = (12,9))
for i in range(len(ET[:])):
    plt.plot(T, ET[i], label = 'N = %d' %(N[i]))
    plt.xscale('log')
    plt.legend(loc ='lower right', prop ={'size':15})
    plt.xlabel("Temperature",size = 15)
    plt.ylabel("Energy at each N site",size = 15)
```

```python
plt.plot(T, ET[6]-ET[5],label="E(8)-E(7)")
plt.plot(T, ET[4]-ET[3],label="E(6)-E(5)")
plt.plot(T, ET[3]-ET[2],label="E(4)-E(3)")
plt.plot(T, ET[2]-ET[1],label="E(2)-E(1)")
plt.xscale('log')
plt.legend(loc = 'best')
plt.xlabel("Temperature",size = 15)
plt.ylabel("Change in Energy at N sites", size = 15)
```
