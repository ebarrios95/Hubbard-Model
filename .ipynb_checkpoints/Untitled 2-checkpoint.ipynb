```python
from numpy import*
from pylab import*
```

```python
i = 1.0*10**-3 #mA
V = 1.0*10**-3 #mV
T = np.arange(100, 300)
#T = 100e3
#e = 1.602 * 10**-19
e =1
Kb = 8.617*10**-5 
```

```python
I = i*(exp((e*V)/(Kb*T)) - 1)
print(I[0])
print(T)
```

```python
i = 1
while True:
    i += 1
    if round(I[0]/2,3) == round(I[i],3):
        print(T[i])
```

```python
plt.plot(T,I)
plt.ticklabel_format(style = 'sci', axis ='x', scilimits = (0,0))
```
